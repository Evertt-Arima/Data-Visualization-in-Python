# Scatter Plots with matplotlib

### 1. Setup environment

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
data = pdr.get_data_yahoo(['GOOG', 'AMZN', 'GLD'], start = '2019-08-01')['Close']
data.head()
```

### 2. Scatter plot of closing prices

```python
plt.scatter(data.GOOG, data.AMZN)
```

### 3. Transform data and create scatter plot

```python
returns = np.log(data).diff()
returns.head()
returns.dropna(inplace=True)
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
plt.axvline(0, c=(.5, .5, .5), ls='--')
plt.axhline(0, c=(.5, .5, .5), ls='--')
plt.style.use('seaborn')
plt.scatter(returns.GOOG, returns.AMZN, c=returns.AMZN, cmap='Spectral', edgecolor='k')
plt.colorbar(orientation='horizontal')
plt.ylim(-.05, .05)
plt.xlim(-.05, .05);
```

### Colormaps matplotlib
* https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html

```python
# default is marker size in points squared --> bubble graph
s = returns['AMZN'] = np.abs(returns.AMZN)
```

```python
plt.scatter(returns.AMZN, returns.GOOG, s=s * 1000)
```

```python
# get more data to demonstrate alpha --> better differentiation of data points
more = pdr.get_data_yahoo(['GOOG', 'AMZN'], start = '2018-01-01')['Close']
more_ret = np.log(more).diff()
more_ret.dropna(inplace=True)
```

```python
plt.scatter(more_ret.GOOG, more_ret.AMZN, alpha=.4)
```
